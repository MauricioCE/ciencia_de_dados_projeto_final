```python
# Importando o pandas
import pandas as pd

df = pd.read_csv("aula_02_dataset_exercicio.csv")
display(df)

mediana_umidade = df["Umidade_Solo"].median()
mediana_co2 = df["Nivel_CO2"].median()

display(f"Mediana da Umidade do Solo: {mediana_umidade}")
display(f"Mediana do Nível de CO2: {mediana_co2}")

df["Umidade_Solo"] = df["Umidade_Solo"].fillna(mediana_umidade)
df["Nivel_CO2"] = df["Nivel_CO2"].fillna(mediana_co2)

display(df)
```
