```python
# Rodar o código apenas uma vez para instalar a biblioteca pandas
# Depois comentar para evitar tentar instalar novamente em execuções futuras
# !pip install pandas

# Importação do pandas para manipulação de dados
import pandas as pd
```

```python
# Carregando dataframe
df = pd.read_csv('aula_02_dataset_1.csv')

# Inspeção visual do dataframe
display(df)

# Verificando o tipo de dados e a presença de valores ausentes
display(df.info())

# Copiando a coluna Temperatura_C para uma nova coluna que irá conter os valores sem NaN
df["Temperatura_C_SEM_NAN"] = df["Temperatura_C"]

# Calculando a mediana da coluna Temperatura_C
# Por que a mediana? Porque é uma medida de tendência central que não é afetada por valores extremos (outliers) e é adequada para substituir valores ausentes em dados numéricos
mediana_temperatura = df['Temperatura_C'].median()
display(f"Mediana: {mediana_temperatura}")

# Substituindo os valores ausentes pela mediana
df = df.fillna({'Temperatura_C_SEM_NAN': mediana_temperatura})
display(df)
display(df.info())
```
