```python
import pandas as pd
```

```python
df = pd.read_csv('aula_02_dataset_1.csv')
display(df)
display(df.info())
```

```python
mediana = df["Temperatura_C"].median()
display(mediana)
```

```python
df = df.fillna({"Temperatura_C": mediana})
display(df)
```
