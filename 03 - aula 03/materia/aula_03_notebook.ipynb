```python
# Carregando dataframe
df = pd.read_csv('aula_03_dataset_1.csv')

# Inspeção visual do dataset
display(df)

# Cuidado!!! No display sai bonito, mas no df está com os espaços em branco
# Inspeção visual de um dos dados com espaços em branco
display("No dataframe tem os espaços em branco:")
display(df["Destino"][0])

# Removendo os espaços em branco
df["Destino"] = df["Destino"].str.strip()
display("Depois de removido os espaços em branco:")
display(df["Destino"][0])
```

```python
# Explo quando os espaços em branco estão no meio do texto
df = pd.read_csv('aula_03_dataset_2.csv')
display(df)
display(df["Destino"][0])
df["Destino"] = df["Destino"].str.strip()
display(df["Destino"][0])
df["Destino"] = df["Destino"].replace(r'\s+', ' ', regex=True)
display(df["Destino"][0])
```
