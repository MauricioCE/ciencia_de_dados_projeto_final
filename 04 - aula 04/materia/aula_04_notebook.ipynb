```python
import pandas as pd
df = pd.read_csv('aula_04_dataset_1.csv')
display(df)
display(df.info())
```

```python
df['Data_Transacao'] = pd.to_datetime(df["Data_Transacao"], format="%d/%m/%Y")
display(df)
display(df.info())
```

```python
df["Valor_RS"] = (
    df["Valor_RS"]
    .str.strip("R$ ")
    .str.replace(".", "", regex=False)
    .str.replace(",", ".", regex=False)
    .astype(float)
)

display(df)
display(df.info())
```
