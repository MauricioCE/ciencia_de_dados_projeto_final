##### Carregando o dataframe

```python
# Carregando dataframe
df = pd.read_csv('aula_04_dataset_1.csv')
```

##### Convertendo datas em formato de string
Problema:

"01/01/2024" está em formato de string, portanto não é reconhecido como data.

Isso pode causar problemas em análises que envolvam datas, como ordenação, filtragem por período, etc.

```python
display(df)
display(df.info()) # Aqui mostra o tipo object (string)
display(df["Data_Transacao"].dtype) # Aqui mostra o tipo dtype('O'). O de object (string)

# Remove espaços em branco, caso existam
df["Data_Transacao"] = df["Data_Transacao"].str.replace(" ", "") 

# Convertendo para o tipo de data do pandas, usando o parâmetro dayfirst=Truepara indicar
# que o formato é dia/mês/ano
df["Data_Transacao"] = pd.to_datetime(df["Data_Transacao"], dayfirst=True)

display(df)
display(df.info())
display(df["Data_Transacao"].dtype)

# Validando
valor_1 = df["Data_Transacao"][0]
valor_2 = df["Data_Transacao"][1]
display(f"Resultado: {valor_1.day + valor_2.day}")
```

##### Convertendo números escritos como string
Problemas:

R$ 2.500,00 tem R$ - Formato string. Precisa remover o R$

2.500,00 - Está no formato brasileiro, mas o Python espera o formato americano (2500.00)

```python
display(df)
display(df.info()) # Aqui mostra o tipo object (string)
display(df["Valor_RS"].dtype) # Aqui mostra o tipo dtype('O'). O de object (string)

# Convertendo
df["Valor_RS"] = (
    df["Valor_RS"]
    .str.strip() # Remove espaços em branco
    .str.replace("R$", "", regex=False) # Remove o R$
    .str.replace(".", "", regex=False) # Remove o ponto dos milhares
    .str.replace(",", ".", regex=False) # Troca a vírgula pelo ponto para o formato americano
    .astype(float) # Converte para float
)

display(df)
display(df.info())
display(df["Valor_RS"].dtype)

# Validando
valor_1 = df["Valor_RS"][0]
valor_2 = df["Valor_RS"][2]
display(f"Resultado: {valor_1 + valor_2}")
```
