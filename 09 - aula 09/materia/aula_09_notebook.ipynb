```python
# Rodar o código apenas uma vez para instalar a biblioteca pandas
# Depois comentar para evitar tentar instalar novamente em execuções futuras
# !pip install pandas

# Importação do pandas para manipulação de dados
import pandas as pd
```

```python
# Carregando dataframe
df = pd.read_csv('aula_09_dataset.csv')
display(df.head())

# Inspeção visual do dataframe
print("\nContagem original da coluna Estado:")
print(df['Estado'].value_counts())
```

```python
# Aplicando a Tríade da Limpeza
# 1. Remover espaços (strip)
# 2. Padronizar caixa (upper)
# 3. Mapear variações (replace)
```

```python
df_limpo = df.copy()

# Passos 1 e 2
df_limpo['Estado'] = df_limpo['Estado'].str.strip().str.upper()

# Passo 3 - Dicionário de mapeamento baseado nos dados
mapeamento = {
    'MINAS GERAIS': 'MG',
    'PARANA': 'PR',
    'CEARA': 'CE',
    'PERNAMBUCO': 'PE',
    'BAHIA': 'BA',
    'S.P.': 'SP'
}

df_limpo['Estado'] = df_limpo['Estado'].replace(mapeamento)

print("\nContagem após a limpeza:")
print(df_limpo['Estado'].value_counts())
```
