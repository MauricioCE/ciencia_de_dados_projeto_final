```python
# Importando o pandas
import pandas as pd

# Abrindo o dataset do exercicio
df = pd.read_csv('aula_10_dataset_exercicio.csv')
display(df)

# Regra de negocio: salario nao pode ser negativo
flag_regra = df['Salario_RS'] < 0

# Metodo estatistico IQR
q1 = df['Salario_RS'].quantile(0.25)
q3 = df['Salario_RS'].quantile(0.75)
iqr = q3 - q1
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr
flag_iqr = (df['Salario_RS'] < limite_inferior) | (df['Salario_RS'] > limite_superior)

# Coluna final de outlier
df['Flag_Outlier'] = flag_regra | flag_iqr
display(df[['ID_Funcionario', 'Salario_RS', 'Flag_Outlier']])

# Comparacao de media, mediana e desvio padrao
estatisticas_antes = df['Salario_RS'].agg(['mean', 'median', 'std']).to_frame().T
estatisticas_antes.index = ['Antes']

df_tratado = df[~df['Flag_Outlier']].copy()
estatisticas_depois = df_tratado['Salario_RS'].agg(['mean', 'median', 'std']).to_frame().T
estatisticas_depois.index = ['Depois']

comparacao = pd.concat([estatisticas_antes, estatisticas_depois])
display(comparacao)

# Dataframe final tratado
display(df_tratado.head(10))
```
