### Aula 01 - Delimitadores diferentes

```python
# Rodar o código apenas uma vez para instalar a biblioteca pandas
# Depois comentar para evitar tentativas desnecessários
# !pip install pandas

# Importação do pandas para manipulação de dados
import pandas as pd
```

```python
# Carregando dataframe
# Errado - O delimitador padrão é ",", mas no dataset é "|"
df_errada = pd.read_csv('aula_01_dataset_1.csv') 
display(df_errada)

# Correto - Especificando o delimitador correto
df_correta = pd.read_csv('aula_01_dataset_1.csv', delimiter='|')
display(df_correta)
```

```python
# Exemplo de dataset com delimitador errado e certo ao mesmo tempo
# Algumas colunas estão separadas por ";" e outras por "|"
df = pd.read_csv('aula_01_dataset_2.csv', delimiter='|')

# Vai printar o datraframe, mas perceba que as duas primeiras colunas aparece com o display, mas não fazem parte do 'df'. 
display(df)

# Para provar isso, basta chamar o método info() do dataframe
display(df.info())

# A forma mais correta de resolver isso é abrir o arquivo CSV em um editor de texto e corrigir o delimitador para que todas as colunas estejam separadas pelo mesmo caractere.
# Se quiser proceeeder, basta usar o método reset_index() do dataframe, passando o parâmetro inplace=True para modificar o dataframe original e drop=True para descartar a coluna de índice antiga.

df.reset_index(inplace=True, drop=True)
display(df)
```
