### Aula 01 - Delimitador não padrão

```python
# Rodar o código apenas uma vez para instalar a biblioteca pandas
# Depois comentar para evitar tentativas desnecessários
# %pip install pandas

# Importação do pandas para manipulação de dados
import pandas as pd
```

```python
# Carregando dataframe
# Errado - O delimitador padrão é ",", mas no dataset é "|"
df_errada = pd.read_csv('aula_01_dataset_1.csv') 
display(df_errada)
```

```python
# Correto - Especificando o delimitador correto
df_correta = pd.read_csv('aula_01_dataset_1.csv', delimiter='|')
display(df_correta)
```
