```python
# Carregando dataframe
df = pd.read_csv('aula_08_dataset.csv')
display(df.head())

# Inspeção visual do dataframe
print("\nTipos originais das colunas:")
print(df.dtypes)
```

```python
# Aplicando o tratamento de datas
# Note que a coluna 'Data_Realizacao' tem formatos mistos (barra, hífen, etc).
# Vamos padronizar usando pd.to_datetime.
```

```python
df_limpo = df.copy()

df_limpo['Data_Realizacao'] = pd.to_datetime(
    df_limpo['Data_Realizacao'], 
    format='mixed', 
    dayfirst=True, 
    errors='coerce'
)

display(df_limpo.head())
print("\nNovo tipo da coluna:")
print(df_limpo['Data_Realizacao'].dtypes)
```

```python
# Extraindo informações úteis
# Agora que é uma data oficial, podemos extrair o Ano e o Mês.
```

```python
df_limpo['Ano'] = df_limpo['Data_Realizacao'].dt.year
df_limpo['Mes'] = df_limpo['Data_Realizacao'].dt.month

display(df_limpo[['Nome_Evento', 'Data_Realizacao', 'Ano', 'Mes']].head())
```
