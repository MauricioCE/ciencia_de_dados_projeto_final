```python
# Importando o pandas
import pandas as pd

# Abrindo o dataset do exercicio
df = pd.read_csv('aula_11_dataset_exercicio.csv')
display(df)

# Separando valor e unidade
df['Valor_Altura'] = df['Altura'].str.extract(r'([\d.]+)').astype(float)
df['Unidade_Altura'] = df['Altura'].str.extract(r'([a-zA-Z]+)$')
df['Valor_Peso'] = df['Peso'].str.extract(r'([\d.]+)').astype(float)
df['Unidade_Peso'] = df['Peso'].str.extract(r'([a-zA-Z]+)$')
display(df[['Produto', 'Altura', 'Unidade_Altura', 'Peso', 'Unidade_Peso']])

# Funcoes de padronizacao
def altura_para_cm(texto):
    valor = float(''.join(ch for ch in texto if ch.isdigit() or ch == '.'))
    unidade = ''.join(ch for ch in texto if ch.isalpha())

    if unidade == 'm':
        return valor * 100
    if unidade == 'cm':
        return valor
    if unidade == 'mm':
        return valor / 10
    return None

def peso_para_kg(texto):
    valor = float(''.join(ch for ch in texto if ch.isdigit() or ch == '.'))
    unidade = ''.join(ch for ch in texto if ch.isalpha())

    if unidade == 'kg':
        return valor
    if unidade == 'g':
        return valor / 1000
    return None

df['Altura_cm'] = df['Altura'].apply(altura_para_cm)
df['Peso_kg'] = df['Peso'].apply(peso_para_kg)
display(df[['Produto', 'Altura', 'Altura_cm', 'Peso', 'Peso_kg']])

# Comparacao de medias por produto
medias_sem_padronizacao = df.groupby('Produto').agg(
    Altura_Media_Sem_Padronizacao=('Valor_Altura', 'mean'),
    Peso_Medio_Sem_Padronizacao=('Valor_Peso', 'mean')
).reset_index()

medias_padronizadas = df.groupby('Produto').agg(
    Altura_Media_cm=('Altura_cm', 'mean'),
    Peso_Medio_kg=('Peso_kg', 'mean')
).reset_index()

comparacao = medias_sem_padronizacao.merge(medias_padronizadas, on='Produto', how='left')
display(comparacao)

# Dataframe final solicitado
df_final = df[['Produto', 'Altura_cm', 'Peso_kg']].copy()
display(df_final)
```
